# Test algorithm with fake data


```python
from noise import noise #import the noise class
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Load the supplied fake data

```python
traces = np.load('fake_noise.npy')
traces_correlated = np.load('fake_noise_correlated.npy')
traces_uncorrelated = np.load('fake_noise_uncorrelated.npy')
```

## Initialize 3 noise instances 
#### one for the combined fake data
#### one for just the correlated traces
#### one for just the un correlated traces

```python
sampleRate = 625e3 #define sample rate
channels = [ 'A','B','C','D'] #define the channel names
fake_noise = noise(traces, sampleRate, channels) #initialize a noise object
fake_noise.set_name('Fake Noise Test') #use the set_name method to name the object

fake_noise_correlated = noise(traces_correlated, sampleRate, channels)
fake_noise_correlated.set_name('Fake Noise correlated')

fake_noise_uncorrelated = noise(traces_uncorrelated, sampleRate, channels)
fake_noise_uncorrelated.set_name('Fake Noise uncorrelated')
```

### Calculate PSDs

```python
fake_noise.calculate_PSD()
fake_noise.calculate_corrCoeff()
fake_noise.calculate_CSD()

fake_noise_correlated.calculate_PSD()
fake_noise_correlated.calculate_corrCoeff()



fake_noise_uncorrelated.calculate_PSD()
fake_noise_uncorrelated.calculate_corrCoeff()
```

### Calculate uncorrelated noise from combined traces

```python
fake_noise.calculate_deCorrelated_noise(verbose = False)
```

## Compare known uncorrelated noise with noise from fit

```python
fake_noise_uncorrelated.plot_PSD(lgc_overlay=True)
fake_noise.plot_deCorrelatedNoise(lgc_overlay=True)

uncorrelated_noise_actual = np.zeros(shape = len(channels))
uncorrelated_noise_fit = np.zeros(shape = len(channels))

uncorrelated_noise_actual_std = np.zeros(shape = len(channels))
uncorrelated_noise_fit_std = np.zeros(shape = len(channels))

for ii in range(len(channels)):
    uncorrelated_noise_actual[ii] = np.mean(fake_noise_uncorrelated.PSD[ii])
    uncorrelated_noise_fit[ii] = np.mean(fake_noise.unCorrNoise[ii])
    uncorrelated_noise_actual_std[ii] = np.std(fake_noise_uncorrelated.PSD[ii])
    uncorrelated_noise_fit_std[ii] = np.std(fake_noise.unCorrNoise[ii])
    
for ii in range(len(channels)):   
    print('Channel: {}'.format(channels[ii]))
    print('    True uncorrelated noise: {:1E} +/- {:1E}'.format(uncorrelated_noise_actual[ii],uncorrelated_noise_actual_std[ii]))
    print('    uncorrelated noise from fit: {:1E} +/- {:1E}'.format(uncorrelated_noise_fit[ii],uncorrelated_noise_fit_std[ii]))
```

## Check correlated noise

```python
correlated_noise_actual = np.mean(fake_noise_correlated.PSD[3])
corr_noise_std = np.std(fake_noise_correlated.PSD[3])
correlated_noise_fit = np.mean(fake_noise.sigC*2)
correlated_noise_fit_std = np.std(fake_noise.sigC*2)

print('True correlated noise: {:1E} +/- {:1E}'.format(correlated_noise_actual,corr_noise_std))
print('correlated noise from fit: {:1E} +/- {:1E}'.format(correlated_noise_fit,correlated_noise_fit_std))
```

# Both the correlated and uncorrelated noise fits agree with data!!!
